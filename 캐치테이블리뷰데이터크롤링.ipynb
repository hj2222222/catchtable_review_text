```python
# !pip install selenium
import pandas as pd
import numpy as np
import time
from selenium import webdriver
from pandas import Series
# !apt-get update
# !apt install chromium-chromedriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless') 
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import chromedriver_autoinstaller
```

```python
names = []
scores = []
contents = []
cates = []
likes = []
comments= []

driver = webdriver.Chrome()
driver.get("https://app.catchtable.co.kr/ct/feed/feedMain")
for j in range(1000):
    driver.execute_script("window.scrollTo(0, 100000000)")
    time.sleep(5)
    content = driver.find_elements(By.CLASS_NAME,'__content')
    score = driver.find_elements(By.CLASS_NAME,'__rating')
    name = driver.find_elements(By.CLASS_NAME,'__name')
    meta = driver.find_elements(By.CLASS_NAME,'__meta')
    like = driver.find_elements(By.CLASS_NAME,'__like')
    comment = driver.find_elements(By.CLASS_NAME,'__comment')
    for i in range(len(content)): 
        names.append(name[i].text)
        scores.append(score[i].text)
        contents.append(content[i].text)
        cates.append(meta[i].text.split('•')[0])
        likes.append(like[i].text)
        comments.append(comment[i].text)
        print(name[i].text)
names_se = Series(data = names)
scores_se = Series(data = scores)
contents_se = Series(data = contents)
cates_se = Series(data = cates)
likes_se = Series(data = likes)
comments_se = Series(data = comments)
data = pd.concat([names_se,scores_se,contents_se,cates_se,likes_se,comments_se],axis=1)
data
```

```python
data = pd.concat([names_se,scores_se,contents_se,cates_se,likes_se,comments_se],axis=1)
data
```

```python
data = data.rename(columns={0:'bistro_name'})
data = data.rename(columns={1:'score'})
data = data.rename(columns={2:'content'})
data = data.rename(columns={3:'category'})
data = data.rename(columns={4:'like'})
data = data.rename(columns={5:'comment'})
data.drop_duplicates(['content'],inplace=True)
```
